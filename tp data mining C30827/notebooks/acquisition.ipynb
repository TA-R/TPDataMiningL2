```python
import sqlite3
import pandas as pd


try:
    # 1. Création d'une base de données locale (fichier : tp_data_mining.db)
    # Cette connexion simule l'accès à un serveur de base de données.
    conn = sqlite3.connect('tp_data_mining.db')

    # 2. Chargement du CSV pour l'intégrer dans la base (Simulation d'importation)
    path_csv = "../data/student-mat.csv"
    df_temp = pd.read_csv(path_csv, sep=';')
    
    # 3. Écriture des données dans une table SQL nommée 'etudiants'
    df_temp.to_sql('etudiants', conn, if_exists='replace', index=False)
    print("Succès : La table 'etudiants' a été créée dans la base SQL locale.")

    # 4. EXTRACRION SQL (Le T.A.F demandé)
    # On exécute une requête SQL pour filtrer les étudiants ayant G3 > 15
    query = "SELECT * FROM etudiants WHERE G3 > 15"
    df_sql_resultat = pd.read_sql(query, conn)

    # 5. Affichage des résultats
    print(f"\nExtraction réussie : {len(df_sql_resultat)} étudiants trouvés.")
    display(df_sql_resultat.head())

except Exception as e:
    print(f"Erreur lors de l'exécution : {e}")

finally:
    # Toujours fermer la connexion pour libérer les ressources
    conn.close()
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# 1. PARTIE WEB SCRAPING (Extraction de citations et auteurs)---------------


# Utilisation d'un site autorisé pour le scraping pédagogique
url_scraping = "http://quotes.toscrape.com/"

try:
    # Envoi de la requête GET au serveur
    response = requests.get(url_scraping)
    
    if response.status_code == 200:
        # Analyse du code HTML de la page
        soup = BeautifulSoup(response.text, 'html.parser')
        
        # Extraction des textes des citations (balises <span> avec classe 'text')
        quotes = [q.get_text() for q in soup.find_all('span', class_='text')]
        
        # Extraction des auteurs (balises <small> avec classe 'author')
        authors = [a.get_text() for a in soup.find_all('small', class_='author')]
        
        # Création d'un DataFrame structuré pour stocker les données web
        df_scraping = pd.DataFrame({
            'Citation': quotes,
            'Auteur': authors
        })
        
        print("--- Source Web Scraping (Quotes) extraite avec succès ---")
        display(df_scraping.head())
    else:
        print(f"Erreur de connexion : Statut {response.status_code}")

except Exception as e:
    print(f"Une erreur est survenue lors du scraping : {e}")



# 2. PARTIE CONSOMMATION D'API (Données JSON) -------------------


# Interroger une API publique pour récupérer des données d'utilisateurs
url_api = "https://jsonplaceholder.typicode.com/users"

try:
    # Récupération de la réponse au format JSON
    api_response = requests.get(url_api)
    
    if api_response.status_code == 200:
        # Transformation du JSON en liste d'objets Python
        data_json = api_response.json()
        
        # Conversion immédiate en DataFrame Pandas
        df_api = pd.DataFrame(data_json)
        
        # Sélection de colonnes spécifiques pour le pipeline final
        df_api_final = df_api[['id', 'name', 'company', 'address']]
        
        print("\n--- Source API (JSONPlaceholder) récupérée avec succès ---")
        display(df_api_final.head())
    else:
        print(f"Erreur API : Statut {api_response.status_code}")

except Exception as e:
    print(f"Erreur lors de l'appel API : {e}")
```
